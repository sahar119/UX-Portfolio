### Weekly report

```python
import pandas as pd
import matplotlib.pyplot as plt
# Reading the CSV file
df = pd.read_csv('Screentime7.csv')
# Converting the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# Defining specific date range
start_date = '2022-08-22'
end_date = '2022-09-02'
# Filtering the DF based on the date range
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
# Calculate statistics on the filtered data
average_totall_attempts = filtered_df["total attempts"].mean()
average_time_spent = filtered_df["Usage"].mean()
prevention_open = filtered_df["Prevention"].mean()
time_saved = prevention_open * average_time_spent
# Print the report
print("Total Attempts To Open:", int(average_totall_attempts), "times")
print("Average Time Spent on Twitter:", int(average_time_spent), "minutes")
print("Preventions:", int(prevention_open), "times")
print("Time Saved:", int(time_saved), "minutes")
# Create a bar chart
categories = ['Total Attempts', 'Average Time Spent', 'Preventions', 'Time Saved']
values = [average_totall_attempts, average_time_spent, prevention_open, time_saved]

plt.bar(categories, values)
plt.ylabel('Values')
plt.title('Statistics')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Screentime7.csv')

# Calculate total attempts
average_totall_attempts = df["total attempts"].mean()
average_totall_attempts = int(average_totall_attempts)

# Calculate Average Time Spent on Twitter (T)
average_time_spent = df["Usage"].mean()
average_time_spent=int(average_time_spent)

# Calculate Prevention  (P)
prevention_open = df["Prevention"].mean()
prevention_open =int(prevention_open)

# Calculate Time Saved using the formula: Time Saved = P * T
time_saved = prevention_open * average_time_spent


print("Your total attemts :", average_totall_attempts, "times")
print("Average Time Spent on Twitter:", average_time_spent, "minutes")
print("Prevention :", prevention_open, "times")
print("Time Saved:", time_saved, "minutes")


# Create a bar chart
categories = ['Total Attempts', 'Average Time Spent', 'Preventions', 'Time Saved']
values = [average_totall_attempts, average_time_spent, prevention_open, time_saved]

plt.bar(categories, values)
plt.ylabel('Values')
plt.title('Statistics')
plt.show()
```
